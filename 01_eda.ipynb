```python

plt.figure(figsize=(6,5))
sns.countplot(data=df,x='Churn',hue='Churn')
plt.title('Customer Churn Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='Tenure', hue='Churn', fill=True, common_norm=False,  alpha=0.5)
plt.title('Density of Tenure by Churn Status')
plt.xlabel('Tenure (Month)', )
plt.ylabel('Density',)
plt.legend(title='Customer Status', labels=['Churned (1)', 'Retained (0)'])
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x='Complain', hue='Churn')

plt.title('Customer Churn vs. Complaints')
plt.xlabel('Did the customer complain?')
plt.ylabel('Number of Customers')
```

```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12, 10))
# Kiszámoljuk a korrelációt és megrajzoljuk
sns.heatmap(numerical_df.corr(), annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, center=0, square=True, linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
columns_to_plot = ['Tenure', 'CashbackAmount', 'DaySinceLastOrder', 'Churn']
pair_plot=sns.pairplot(df[columns_to_plot], hue='Churn',  corner=True)
pair_plot.figure.suptitle('Pairwise Relationships of Key Features by Churn', y=1.02, )
plt.show()
```
