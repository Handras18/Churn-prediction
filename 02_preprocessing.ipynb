```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_excel("data/E Commerce Dataset.xlsx",sheet_name="E Comm")
data_dict=pd.read_excel("data/E Commerce Dataset.xlsx",sheet_name="Data Dict")
```

```python
df_numeric=df.select_dtypes(include=['number']).columns.to_list()
df_categorical=df.select_dtypes(include=['str']).columns.to_list()
```

```python
df_numeric
df_categorical
```

```python
for col in df_numeric:
    if df[col].isnull().sum() > 0:
        df[f'{col}_is_missing'] = df[col].isnull().astype(int)
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
```

```python
df.isnull().sum()
```

```python
for col in df_categorical:
    uniques=df[col].unique()
    print(f'{col} ({len(uniques)} unique values):\n  {uniques}\n')
```

```python
df['PreferredLoginDevice']=df['PreferredLoginDevice'].str.replace('Mobile Phone','Phone',regex=False)
df['PreferredPaymentMode']=df['PreferredPaymentMode'].str.replace('Credit Card','CC',regex=False)
df['PreferredPaymentMode']=df['PreferredPaymentMode'].str.replace('Cash on Delivery','COD',regex=False)
df['PreferedOrderCat']=df['PreferedOrderCat'].str.replace('Mobile Phone','Phone',regex=False)
```

```python
df[['PreferredLoginDevice','PreferredPaymentMode','Gender','PreferedOrderCat','MaritalStatus']]
```

```python
encoder=OneHotEncoder(sparse_output=False,drop='first')
encoded_list=encoder.fit_transform(df[df_categorical])
encoded_df = pd.DataFrame(encoded_list, 
                          columns=encoder.get_feature_names_out(df_categorical))

df=pd.concat([df,encoded_df],axis=1)
df = df.drop(columns=df_categorical)
```
